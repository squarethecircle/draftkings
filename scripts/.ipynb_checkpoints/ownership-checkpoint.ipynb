```python
import os
import zipfile
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import statsmodels.api as sm
import statsmodels.stats.api as sms
import scipy.stats as stats
import pandas as pd
import sys
sys.path.append('..')
from constants import *

CUR_YEAR = 2015
CUR_WEEK = 9

hf = pd.concat([pd.read_pickle('../data/histdata').dropna(), pd.read_pickle('../data/curprojs')])
hf['Opp']=hf.apply(lambda r: r['Matchup'].split(' ')[2] if isinstance(r['Matchup'],str) else 'NA',axis=1)
sals = pd.read_csv('../data/DKSalariesCurrent.csv')
sals['PID'] = sals.apply(lambda r: generate_pid(r['Name'], r['Position']),axis=1)
sals['Week'] = CUR_WEEK
sals['Year'] = CUR_YEAR
hf = hf.merge(sals[['PID', 'Week', 'Year', 'Salary']], on=['PID', 'Week', 'Year'], how='left')
hf['Salary_x'] = hf.apply(lambda r: r['Salary_x'] if not np.isnan(r['Salary_x']) else r['Salary_y'], axis=1)
hf.drop('Salary_y', axis=1)
hf.rename(columns={'Salary_x':'Salary'}, inplace=True)
pids = hf['PID'].unique()
nums = dict([(pid, hf[hf['PID']==pid].shape[0]) for pid in pids])

projs=pd.read_csv(os.path.expanduser('~/Dropbox/DFS/data/contests/Week{}/playervals.csv'.format(CUR_WEEK)))
projs.columns=['PID','points']
hf1=hf[hf['Week']==CUR_WEEK][hf['Year']==CUR_YEAR].merge(projs, on=['PID'])[['PID','Pos','Team','Opp','points', 'Avg Rank']]
hf1=hf1.merge(sals[['PID', 'Salary']], on=['PID'])
hf1['PPD'] = hf1.apply(lambda r: r['points'] / r['Salary'], axis=1)

def conv_player_string(player_str):
    delim_strs = ['QB', 'RB', 'WR', 'TE', 'FLEX', 'DST']
    splits = player_str.split(' ')
    players_list = []
    if player_str.strip() == '':
        return {}
    for s in splits:
        if s in delim_strs:
            players_list.append([s, []])
        else:
            players_list[-1][1].append(s)
    ret = []
    for pos, nm in players_list:
        nm = ' '.join(nm)
        if pos != 'FLEX':
            ret.append(generate_pid(nm, pos))
        else:
            candidates = [generate_pid(nm, 'RB'), generate_pid(nm, 'WR'), generate_pid(nm, 'TE')]
            for c in candidates:
                if hf1[hf1['PID']==c].shape[0] != 0:
                    ret.append(c)
                    break
    return ret

dn = os.path.expanduser('~/Dropbox/DFS/data/contests/')
contests_by_week = {}
pvals_by_week = {}
for w in [w for w in os.listdir(dn) if os.path.isdir(os.path.join(dn, w))]:
    contests_by_week[w] = {}
    with open(os.path.join(dn, w, 'playervals.csv'), 'r') as re:
        pvals_by_week[w] = dict((' '.join(x[0].split('_')[:-1]), x[1]) for x in
                map(lambda l: l.strip().split(','), re.readlines()))
    for zf in [f for f in os.listdir(os.path.join(dn, w)) if 'zip' in f]:
        z = zipfile.ZipFile(os.path.join(dn, w, zf))
        for f in z.namelist():
            if f[0] == '_':
                continue
            contest_id = f.split('.')[0].split('-')[2]
            with z.open(f) as re:
                contests_by_week[w][contest_id] = \
                        map(lambda l: l.strip().split(','), re.readlines()[1:])
                #for entry in contests_by_week[w][contest_id]:
                #    entry[2] = ' '.join(entry[2].split(' ')[:-1])
                #    entry[5] = conv_player_string(entry[5])
contests = contests_by_week['Week7']
for c in contests:
    contests[c] = map(lambda lis: conv_player_string(lis[5]), contests[c])
```

```python
pids = {}
num_lineups = 0.
for c in contests:
    for lineup in contests[c]:
        num_lineups += 1
        for pid in lineup:
            if pid not in pids:
                pids[pid] = 0
            pids[pid] += 1
for i, r in hf1.iterrows():
    hf1.set_value(i, 'PctOwned', pids[r['PID']]/num_lineups if r['PID'] in pids else 0)
```

```python
pos='WR'
toplot = hf1[hf1['Pos']==pos] if pos is not None else hf1
plt.scatter(toplot['PPD'], toplot['PctOwned'])
plt.xlim((0.001,0.0045))
plt.show()
```
